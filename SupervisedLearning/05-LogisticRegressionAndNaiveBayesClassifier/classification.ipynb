```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('Cleaned_Algerian_forest_fires_dataset.csv')
df
```

Classes: 1: NotFire, 0: Fire

---

```python
df.Classes.value_counts()
```

```python
plt.pie(df.Classes.value_counts(),labels=['Fire', 'Not Fire'],autopct='%1.1f%%')
```

```python
data=df.drop(['day','month','year'],axis=1)
data
```

```python
data.corr()
```

```python
data.drop(['Ws','Rain','Region'],axis=1,inplace=True)
data
```

```python
x=data.drop('Classes',axis=1)
y=data.Classes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=24)
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
from sklearn.metrics import confusion_matrix,fbeta_score,precision_score,recall_score
```

```python
model.fit(x_train_scaled,y_train)
y_pred=model.predict(x_test_scaled)
```

```python
model.predict_proba(x_test_scaled)
```

```python
model.score(x_test_scaled,y_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
fbeta_score(y_test,y_pred,beta=1)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
auc=roc_auc_score(y_test,model.predict_proba(x_test_scaled)[:,1])
auc
```

```python
fpr,tpr,threshold=roc_curve(y_test,model.predict_proba(x_test_scaled)[:,1])
fpr,tpr,threshold
```

```python
plt.plot([0,1],[0,1])
plt.plot(fpr,tpr)
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'C':[1,1.2,1.5,1.7,2.3,2.7],
        'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
        'penalty': ['l1', 'l2', 'elasticnet']
        }
grid=GridSearchCV(model,param_grid=params,cv=3)
```

```python
grid.fit(x_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params={'C':[1,1.2,1.5,1.7,2.3,2.7],
        'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
        'penalty': ['l1', 'l2', 'elasticnet']
        }
rs=RandomizedSearchCV(model,param_distributions=params,cv=3)
```

```python
rs.fit(x_train_scaled,y_train)
```
