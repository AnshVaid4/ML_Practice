```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn.datasets import load_iris
data=load_iris()
data
```

```python
print(data.DESCR)
```

```python
x=pd.DataFrame(data.data,columns=['SL','SW','PL','PW'])
x
```

```python
y=data.target
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc_pred=dtc.predict(x_test)
```

```python
dtc.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,dtc_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,dtc_pred))
```

```python
from sklearn import tree
plt.subplots(figsize=(20,10))
tree.plot_tree(dtc,filled=True)
```

# Post Pruning<br>
After looking at graph, the 2nd node itself is sufficient, instead of further splits

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(max_depth=2)
dtc.fit(x_train,y_train)
dtc_pred=dtc.predict(x_test)
```

```python
dtc.score(x_test,y_test)
```

```python
print(classification_report(y_test,dtc_pred))
```

```python
plt.subplots(figsize=(5,4))
tree.plot_tree(dtc,filled=True)
```

```python
confusion_matrix(y_test,dtc_pred)
```

# Pre Pruning<br>
Which hyperparameter should best suit the model

```python
params={'criterion':['gini', 'entropy', 'log_loss'],
        'splitter':['best', 'random'],
        'max_features':[2,3,4],
        'max_depth':[1,2,3,4,5,6],
        'min_samples_split':[2,3,4],
        'random_state':[5,7,13,17]}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(dtc,param_grid=params)
grid.fit(x_train,y_train)
grid_pred=grid.predict(x_test)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
print(classification_report(y_test,grid_pred))
```

```python
confusion_matrix(y_test,grid_pred)
```
