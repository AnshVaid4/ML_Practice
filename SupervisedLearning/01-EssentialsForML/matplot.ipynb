```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[5,3,1,7,9]
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line chart')
plt.xticks([2,4,6])
plt.yticks([1,2,3,4,5,6,7,8,9])
plt.show()
```

```python
x=[1,2,3,4,5]
y=[5,3,1,7,9]
plt.plot(x,y, color='red', linestyle=':', marker='o', linewidth=2, markersize=4)
plt.grid()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line chart')
plt.xticks([2,4,6])
plt.yticks([1,2,3,4,5,6,7,8,9])
plt.show()
```

```python
np.random.seed(20)
x=np.random.randint(20,size=10)
y1=np.random.randint(20,size=10)
y2=np.random.randint(20,size=10)
y3=np.random.randint(20,size=10)
y4=np.random.randint(20,size=10)

plt.figure(figsize=(15,5))  # (width, height)
p1=plt.subplot(2,2,1)       # 2,2 => 4 charts, 1,2,3,4 1st column, 2nd column, 3rd column, 4th column
p2=plt.subplot(2,2,2)
p3=plt.subplot(2,2,3)
p4=plt.subplot(2,2,4)

p1.plot(x,y1)
p2.scatter(x,y2)
p3.pie(x)
p4.plot(y1,y2)

plt.show()
```

```python
category=['A','B','C','D']
value=[10,40,20,60]
plt.bar(category,value, color='green')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar graph')
plt.show()
```

```python
x=np.random.randn(120)
print(x,x.min(),x.max())
plt.hist(x,bins=12)
```

```python
category=['A','B','C','D']
value=[10,40,20,60]
plt.pie(value, labels=category, autopct="%i")
# plt.title('Bar graph')
plt.show()
```
