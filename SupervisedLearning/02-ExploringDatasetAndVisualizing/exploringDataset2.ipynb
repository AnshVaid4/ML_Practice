```python
df['dateDay']=df.Date_of_Journey.apply(lambda x: x.split('/')[0])
df['dateMonth']=df.Date_of_Journey.apply(lambda x: x.split('/')[1])
df['dateYear']=df.Date_of_Journey.apply(lambda x: x.split('/')[2])
df.head()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
df.head()
```

```python
df['DepHr']=df.Dep_Time.apply(lambda x: x.split(':')[0])
df['DepMin']=df.Dep_Time.apply(lambda x: x.split(':')[1])
df['ArrHr']=df.Arrival_Time.apply(lambda x: x.split(':')[0])
df['ArrMin']=df.Arrival_Time.apply(lambda x: x.split(':')[1].split(' ')[0])
df.head()
```

```python
df.drop('Arrival_Time',axis=1, inplace=True)
df.drop('Dep_Time',axis=1, inplace=True)
df.head()
```

```python
def cleanHr(x):
    x=x.replace('m','').replace('h','').split(' ')[0]
    return x

def cleanMin(x):
    try:
        x=x.replace('m','').replace('h','').split(' ')[1]
    except:
        x=0
    return x


df['DurationHr']=df.Duration.apply(lambda x: cleanHr(x))
df['DurationMin']=df.Duration.apply(lambda x: cleanMin(x))
df.head()
```

```python
df.drop('Duration', axis=1, inplace=True)
df.drop('Route', axis=1, inplace=True)
df.head()
```

```python
df.Additional_Info.unique()
```

```python
df.Total_Stops.unique(), df.Total_Stops.value_counts(), df.Total_Stops.fillna('1 stop',inplace=True)
```

```python
df.Total_Stops.value_counts()
```

```python
df.Total_Stops=df.Total_Stops.map({'1 stop':1, '2 stops':2, '3 stops':3 , '4 stops':4, 'non-stop':0})
df.head()
```

```python
df.dtypes
```

```python
df.Total_Stops=df.Total_Stops.astype(int)
df.dateDay=df.dateDay.astype(int)
df.dateMonth=df.dateMonth.astype(int)
df.dateYear=df.dateYear.astype(int)
df.DepHr=df.DepHr.astype(int)
df.DepMin=df.DepMin.astype(int)
df.ArrHr=df.ArrHr.astype(int)
df.ArrMin=df.ArrMin.astype(int)
df.DurationHr=df.DurationHr.astype(int)
df.DurationMin=df.DurationMin.astype(int)
df.dtypes
```
