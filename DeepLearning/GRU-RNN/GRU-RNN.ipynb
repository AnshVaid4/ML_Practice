```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from sklearn.model_selection import train_test_split
import nltk
```

```python
with open('hamlet.txt','r') as file:
    text=file.read().lower()
text
```

```python
t=Tokenizer()
t.fit_on_texts([text])
print(len(t.word_index))
t.word_index
```

```python
with open('hamlet.txt','r') as file:
    text_lines=file.readlines()
for i in text_lines:
    if i=='\n':
        text_lines.remove('\n')
text_lines
```

```python
import re
for i in range(len(text_lines)):
    text_lines[i]=re.sub("(^\s*)|(\n)",'',text_lines[i]).lower()
text_lines_copy=text_lines
text_lines
```

```python
[text_lines[0]],t.texts_to_sequences([text_lines[0]])
```

```python
possibilities=[]
for i in range(len(text_lines)):
    token=t.texts_to_sequences([text_lines[i]])[0]
    for i in range(2,len(token)):
        possibilities.append(token[:i])
possibilities
```

```python
max_length=0
for i in possibilities:
    max_length=max(max_length,len(i))
max_length
```

```python
for i in range(len(possibilities)):
    possibilities[i]=pad_sequences([possibilities[i]],padding='pre',maxlen=max_length)[0]
possibilities
```

```python
possibilities=np.array(possibilities)
possibilities
```

```python
x=possibilities[:,:-1]
y=possibilities[:,-1]
```

```python
x
```

```python
y
```

```python
x.shape,y.shape
```

```python
y=tf.keras.utils.to_categorical(y,num_classes=len(t.word_index))
y
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=19)
```

```python
x_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense, Dropout
```

```python
model=Sequential()
model.add(Embedding(len(t.word_index)+1,100,input_shape=(12,)))
model.add(GRU(120, return_sequences=True))
model.add(Dropout(0.2))
model.add(GRU(80))
model.add(Dropout(0.1))
model.add(Dense(len(t.word_index),activation="softmax"))
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
#es=EarlyStopping(monitor='val_loss',patience=15,restore_best_weights=True)
model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test),verbose=1)
```
