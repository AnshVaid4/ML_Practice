```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn.datasets import load_iris
data=load_iris()
x=pd.DataFrame(data.data,columns=data.feature_names)
y=data.target
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
x
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x)
x_pca
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1],c=y)
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
plt.subplots(figsize=(15,5))
dendrogram(linkage(x_pca,method='median'))
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
plt.subplots(figsize=(15,5))
dendrogram(linkage(x_pca,method='ward'))
plt.plot([0,9999],[20,20],color='red',linestyle=':')    # K=2 since this line cuts 2 vertical lines
plt.show()
```

---
<br><br>
K=2 clusters for **Hirarchical Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=2,linkage='ward')
model.fit(x_pca)
model.labels_
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1],c=model.labels_)
```
