```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=780,n_features=2,centers=2,cluster_std=1,random_state=3)
sb.scatterplot(x=x[:,0],y=x[:,1],hue=y)
```

```python
x
```

```python
x=np.resize(x,(781,2))
y=np.insert(y,780,1)
```

```python
x[780][:]=np.array([[-6,8]],ndmin=2)   #Global outlier
y[780]=1
```

```python
x[780][:]
```

```python
sb.scatterplot(x=x[:,0],y=x[:,1],hue=y)
```

```python
from sklearn.neighbors import LocalOutlierFactor
model=LocalOutlierFactor(n_neighbors=5,algorithm='ball_tree',contamination=0.011)
preds=model.fit_predict(x)
```

```python
sb.scatterplot(x=x[:,0],y=x[:,1],hue=preds)
```
