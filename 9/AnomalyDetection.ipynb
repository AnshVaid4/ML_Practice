```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=780,n_features=2,centers=2,cluster_std=1.85,random_state=2)
sb.scatterplot(x=x[:,0],y=x[:,1],hue=y)
```

```python
x
```

```python
from sklearn.ensemble import IsolationForest
outlier=IsolationForest(n_estimators=100,contamination=0.4)
outlier.fit(x)
```

```python
preds=outlier.predict(x)
preds
```

```python
sb.scatterplot(x=x[:,0],y=x[:,1],hue=preds)
```
