`Churn_Modelling.csv` is used for dataset, which has various fields and the result showing if customer exited the bank or not

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
df=pd.read_csv('Churn_Modelling.csv').drop('RowNumber',axis=1)
df.head()
```

```python
df.drop(['CustomerId','Surname'],axis=1,inplace=True)
df
```

```python
pd.get_dummies(df['Geography'],dtype='int')
```

```python
for column in pd.get_dummies(df['Geography'],dtype='int').columns:
    df[column]=pd.get_dummies(df['Geography'],dtype='int')[column]
df.drop('Geography',axis=1)
```

```python
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df # 0:Female, 1:Male
```

```python
x=df.drop(['Exited','Geography'],axis=1)
y=df['Exited']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)
len(x_train),len(x_test),len(y_train),len(y_test)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train[0]
```

# ANN with Tensorflow

<br><br>

## Steps:

- Sequential network
- Dense (Hidden neuron)
- Activation function (sigmoid, softmax, tanh, relu, leaky relu)
- Optimizer (Back propagation to update weights)
- Loss function (Get the minimal value in gradient discent)
- Metrics

<br><br><br><br>

### Classification Problem

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
#input data
x_train.shape #i.e. 12
```

```python
model=Sequential(
    [
        Dense(64, activation='relu',input_shape=(12,)), # Input layer connected to 1st hidden layer
        Dense(32, activation='relu'), # Hidden layer 2
        Dense(1, activation='sigmoid') # Output layer
    ]
)

# (12*64)+(64*32)+(32*1)+(64+32+1) => 2945 parameters
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
lo=tf.keras.losses.BinaryCrossentropy()

model.compile(optimizer=opt,loss=lo,metrics=['accuracy'])
```

```python

tensorflow_callback=TensorBoard(log_dir='logs/',histogram_freq=1)
```

```python
early_stopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping]
)
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('Churn_Modelling.csv').drop('RowNumber',axis=1)
df.drop(['CustomerId','Surname'],axis=1,inplace=True)
for column in pd.get_dummies(df['Geography'],dtype='int').columns:
    df[column]=pd.get_dummies(df['Geography'],dtype='int')[column]
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
x=df.drop(['Exited','Geography'],axis=1)
y=df['Exited']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)
len(x_train),len(x_test),len(y_train),len(y_test)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
def create_model(hidden_units,hidden_units1):
    model = Sequential()
    model.add(Dense(hidden_units, activation='relu', input_dim=(x_train.shape[1])))
    model.add(Dense(hidden_units1, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model


model = KerasClassifier(model=create_model, verbose=0)
ann_param_grid = {'model__hidden_units': [32,64,128],
                  'model__hidden_units1': [16,32,64],
                  'epochs': [100],
                  'batch_size':[16,32]}

grid_search = GridSearchCV(estimator=model, param_grid=ann_param_grid, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
accuracy_score(y_test,y_pred)
```
