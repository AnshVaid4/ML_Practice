```python
import nltk
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('spam_ham_dataset.csv')
data
```

```python
data.drop(['Unnamed: 0','label_num'],axis=1,inplace=True)
```

```python
data
```

```python
import re
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
x=data.text
y=data.label
```

```python
re.sub('[^a-zA-Z]',' ',x[0])
```

```python
word_tokenize(re.sub('[^a-zA-Z]',' ',x[0]).lower())
```

```python
''' 
Transforming x by apply regular expression, lowering the case, tokenizing the sentence to words and checking for
stopwords. If word is not a stopword then send it for lemmatization and then the whole row is transformed by joining
the resultant words in a sentence and rewriting the row index. Similarly for each row.
'''
wnl=WordNetLemmatizer()
for i in range(len(x)):
    sentence=[]
    words=word_tokenize(re.sub('[^a-zA-Z]',' ',x[i]).lower())
    for word in words:
        if word in stopwords.words('english'):
            words.remove(word)
        else:
            word=wnl.lemmatize(word=word, pos='v')
            sentence.append(word)
            
    sentence=' '.join(sentence)
    x[i]=sentence
```
