```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv('tour_package.csv')
data.drop('CustomerID',axis=1,inplace=True)
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
for column in data.columns:
    print(column,':',data[column].unique())
```

```python
data.Gender.replace('Fe Male','Female',inplace=True)
data.MaritalStatus.replace('Single','Unmarried',inplace=True)
```

```python
cols=['TypeofContact','CityTier','Occupation','Gender','ProductPitched','MaritalStatus','Designation']
for column in cols:
    data[column]=data[column].astype('category')
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
data.Age.fillna(int(np.mean(data.Age)),inplace=True)
data.TypeofContact.fillna(data.TypeofContact.mode()[0],inplace=True)
data.DurationOfPitch.fillna(int(np.mean(data.DurationOfPitch)),inplace=True)
data.NumberOfFollowups.fillna(int(np.mean(data.NumberOfFollowups)),inplace=True)
data.PreferredPropertyStar.fillna(np.mean(data.PreferredPropertyStar),inplace=True)
data.NumberOfTrips.fillna(int(np.mean(data.NumberOfTrips)),inplace=True)
data.NumberOfChildrenVisiting.fillna(int(np.mean(data.NumberOfChildrenVisiting)),inplace=True)
data.MonthlyIncome.fillna(np.mean(data.MonthlyIncome),inplace=True)
```

```python
data.isna().sum()
```

```python
data
```

```python
cols=['TypeofContact','CityTier','Occupation','Gender','ProductPitched','MaritalStatus','Designation']
for column in cols:
    print(column,':',data[column].unique())
```

```python
codes=[{'Self Enquiry':0, 'Company Invited':1},
       {'Salaried':0, 'Free Lancer':1, 'Small Business':2, 'Large Business':3},
       {'Female':0, 'Male':1},
       {'Deluxe':0, 'Basic':1, 'Standard':2, 'Super Deluxe':3, 'King':4},
       {'Unmarried':0, 'Divorced':1, 'Married':2},
       {'Manager':0, 'Executive':1, 'Senior Manager':2, 'AVP':3, 'VP':4}]

cols=['TypeofContact','Occupation','Gender','ProductPitched','MaritalStatus','Designation']

for i in range(len(cols)):
    data[cols[i]].replace(codes[i],inplace=True)
```

```python
data
```

```python
x=data.drop('ProdTaken',axis=1)
y=data.ProdTaken
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=23)
x_train.shape,x_test.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
rfc=RandomForestClassifier()
dtc=DecisionTreeClassifier()
lr=LogisticRegression()
models=[('Random Forest Classifier',rfc),('Decision Tree Classifier',dtc),('Logistic Regression',lr)]
```

```python
y.value_counts()
```

```python
for model in models:
    m=model[1]
    m.fit(x_train_scaled,y_train)
    pred=m.predict(x_test_scaled)

    print(model[0],'score:',m.score(x_test_scaled,y_test))
    print(model[0],'confusion metrics:',confusion_matrix(y_test,pred))
    print(model[0],'classification report:',classification_report(y_test,pred))
    print('-------------------------------------------\n')
    
```

<br><br><br>
### Since Random forest has highest score

```python
params={'criterion':['gini', 'entropy', 'log_loss'],
        'max_depth':[10,15,20,25,30],
        'max_leaf_nodes':[2,4,6],
        'class_weight':[{0: 1, 1: 1.5}, {0: 1, 1: 2}, {0: 1, 1: 2.5}]}

from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(rfc,param_grid=params,cv=5)
grid.fit(x_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid_pred=grid.predict(x_test_scaled)
print(classification_report(y_test,grid_pred))
```
