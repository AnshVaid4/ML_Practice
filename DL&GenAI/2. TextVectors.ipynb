# Bag of Words

```python
import pandas as pd
df=pd.read_csv('spam_ham_dataset.csv').drop('Unnamed: 0',axis=1)
df=df[:500]
```

```python
df.head()
```

```python
# data cleaning and preprocessing
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

stemmer=SnowballStemmer('english')
```

```python
corpus=[]
for i in range(len(df)):
    text=df.iloc[i]['text'].strip()
    retext=re.sub('[^a-zA-Z]',' ',text)
    retext=retext.lower()
    words=[]
    for word in retext.split(' '):
        if word not in stopwords.words('english') and word != '':
            word=stemmer.stem(word)
            words.append(word.strip())
    text=' '.join(words)
    df.iloc[i]['text']=text
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
model=CountVectorizer(max_features=500) # Top 500 frequent words are used
```

```python
x=model.fit_transform(df['text'][:350])
x.toarray()
```

```python
x[4].toarray()
```

```python
model.vocabulary_
```

```python
model=CountVectorizer(max_features=500,ngram_range=(2,3)) # Top 500 frequent words are used
```

```python
x=model.fit_transform(df['text'][:350])
x.toarray()
```

```python
x[4].toarray()
```

```python
model.vocabulary_
```

# Word2Vec

```python
from gensim.models import Word2Vec

sentences = [["the", "cat", "sat", "on", "the", "mat"],
             ["the", "dog", "barked"],
             ["the", "dog", "sat", "on", "the", "rug"],
             ['the','sky','is','blue']]

# Training Word2Vec on our sentences
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4) 

vector = model.wv['the']
print(vector)
```

```python
model.predict_output_word(['the', 'cat', 'sat','on'], topn=5)
```

```python
model.predict_output_word(['the', 'sky', 'is'], topn=15)
```

```python
model2 = Word2Vec([['the']], vector_size=100, window=1, min_count=1, workers=4)
model2.wv['the']
```

```python
model2.predict_output_word(['sky', 'is'], topn=3)
```

```python
model2.predict_output_word(['the','sky', 'is'], topn=3)
```

```python
model2.wv.most_similar('the', topn=5)
```
