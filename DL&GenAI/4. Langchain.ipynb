# Creating index and adding embeddings to elasticsearch

```python
from elasticsearch import Elasticsearch
es = Elasticsearch([{'scheme': 'http', 'host': '192.168.25.135', 'port': 9200}])
es
```

```python
index_name = "test"
vector_dimension= 1024

mapping = {
    "mappings": {
        "properties": {
            "doctext": {"type": "text"},
            "embeddings": {"type": "dense_vector", "dims": vector_dimension},
        }
    }
}

createreferences=0
# Create the index
if not es.indices.exists(index=index_name):
    createreferences=1
    es.indices.create(index=index_name, body=mapping)
else:
    print(f"Index '{index_name}' already exists!")
```

```python
import ollama
oe=OllamaEmbeddings(model='mxbai-embed-large')   # When we wanna use ollama embedding model using LangChain

if createreferences:
    for doc in documentsSplit:
        text=doc.page_content
        embeddings=ollama.embeddings(model='mxbai-embed-large',prompt=text)
        document = {
        "doctext": text,
        "embeddings": embeddings.embedding
        }
        response = es.index(index=index_name, body=document)
        print(response)
```

```python
response = es.search(index=index_name, body={
    "query": {
        "match": {
            "doctext": "Bad"
        }
    }
})
response,len(response['hits']['hits']),response['hits']['hits']
```

```python
e=ollama.embeddings(model='mxbai-embed-large',prompt="crispy all the way through")

response = es.search(index=index_name, body={
    "size": 4,"track_total_hits": True,
    "query": {
        "script_score": {
            "query":{
        "match_all": { }
            },
            "script":{
                "source": "cosineSimilarity(params.query_vector, 'embeddings')", # cosineSimilarity between params.query_vector & respective 'embeddings' field in documents
                "params": {
                    "query_vector": e.embedding  
                }
            }
    }
    }
})

print(response,len(response['hits']['hits']),'\n')
print(response['hits']['hits'][0]['_source']['doctext'],'\n')
print(response['hits']['hits'][1]['_source']['doctext'],'\n')
print(response['hits']['hits'][2]['_source']['doctext'],'\n')
print(response['hits']['hits'][3]['_source']['doctext'],'\n')
```

```python
from langchain_ollama import OllamaLLM
# model = OllamaLLM(model="llama3.2:1b")  # This Model doesn't support Tools so it won't work
model = OllamaLLM(model="llama3.2:3b")  # This Model supports Tools, so it generates the results correctly
# model = OllamaLLM(model="deepseek-r1:1.5b")
```

```python
from langchain_core.prompts import ChatPromptTemplate

def getContext(string):
    embedding=ollama.embeddings(model='mxbai-embed-large',prompt=string)
    results= es.search(index=index_name, body={
    "size": 5,"track_total_hits": True,
    "query": {
        "script_score": {
            "query":{
        "match_all": { }
            },
            "script":{
                "source": "cosineSimilarity(params.query_vector, 'embeddings')", # cosineSimilarity between params.query_vector & respective 'embeddings' field in documents
                "params": {
                    "query_vector": embedding.embedding 
                }
            }
    }
    }
    })
    
    context=[]
    for i in range(5):
        context.append(results['hits']['hits'][i]['_source']['doctext'])
    return context


def getContextWrapper(question):

    results=getContext(question)
    results=', Other Context: '.join(results)

    template = """
        You are an food reviewer who likes to pricisely answer to queries of people about the places based on the existing reviews.
        The review is as follows: {context}

        Question: {question}
        """

    prompt = ChatPromptTemplate.from_template(template)

    chain = prompt | model
    response=chain.invoke({'context':results,'question':question})

    return response


from langchain.tools import Tool

rag_tool = Tool(
    name="RAGSearch",
    func=getContextWrapper,
    description="Useful for retrieving internal restaurant review knowledge from embedded CSV data"
)
```

```python
from langchain.agents import initialize_agent, AgentType




agent = initialize_agent(
    tools=[rag_tool],
    llm=model,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    handle_parsing_errors=True
)


response = agent.run("Does the shop have crispy pizzas?")
response
```
